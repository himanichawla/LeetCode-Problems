```python
def checkInteger(str):
    n = len(str)
    list= [0] * n
    for i, chr in enumerate(str):
        if chr.isdigit():
            list[i] = 1
    for i in list:
        if i==0:
            return False
    return True
```

Find largest word in dictionary by deleting some characters of given string

```python
def largestword(dic , str1):
    count = [0] * 26
    for char in str1:
        count[ord(char)- ord("a")] +=1
    maxlength  =0
    maxword = " "
    for word in dic:
        length =0
        for char in word:
            if count[ord(char)- ord("a")] > 0:
                length +=1
        if (length == len(word) and length > maxlength):
            maxlength = length
            maxword = word
    return maxword   
            
```
